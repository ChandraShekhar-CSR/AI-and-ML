## Exercise - 3

1. Load and Prepare Data:
• Fetch an image of you choice.{If colour convert to grayscale}
• Center the dataset - Standaridze the Data.
• Calculate the covaraince matrix of the Standaridze data.

```python
data = gray_image.copy()
plt.imshow(data, cmap='gray')
plt.axis('off')
plt.title("Data")
plt.show()
```

```python
data.shape
```

```python
mean=np.mean(data,axis = 0)
data_centered= data - mean
data_centered.shape
```

```python
#covariance matrix
covariance_matrix = np.cov(data_centered, rowvar=False)
# to see how two features change together
print(covariance_matrix.shape)
covariance_matrix
```

2. Eigen Decomposition and Identifying Principal Components:
• Compute Eigen Values and Eigen Vectors.
• Sort the eigenvalues in descending order and choose the top k eigenvectors corresponding to
the highest eigenvalues.
• Identify the Principal Components with the help of cumulative Sum plot.

```python
eigen_values, eigen_vectors = np.linalg.eigh(covariance_matrix)
#sort the indices in deacreasing order
sorted_indices = np.argsort(eigen_values)[::-1]
#sort the indices in descending order
eigen_values = eigen_values[sorted_indices]
eigen_vectors = eigen_vectors[:, sorted_indices]
```

```python
explained_variance_ratio = eigen_values / np.sum(eigen_values)
plt.plot(np.cumsum(explained_variance_ratio))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance Plot')
plt.show()
```

3. Reconstruction and Experiment:

• Reconstruction: Transform the original data by multiplying it with the selected eigenvec-
tors(PCs) to obtain a lower-dimensional representation.

• Experiments: Pick Four different combination of principal components with various ex-
plained variance value and compare the result.

• Display the Results and Evaluate.

```python
k = 35
components = eigen_vectors[:,:k]
```

```python
compressed_data=np.dot(data_centered,components)
```

```python
decompressed_data=np.dot(compressed_data,components.T)+mean
```

```python
#original image
plt.subplot(1, 3, 1)
plt.imshow(data, cmap='gray')
plt.axis('off')
plt.title('Original Image')

#compressed image
plt.subplot(1, 3, 2)
plt.imshow(compressed_data, cmap='gray',aspect='auto')
plt.axis('off')
plt.title('Compressed Image')

#decompressed image
plt.subplot(1, 3, 3)
plt.imshow(decompressed_data, cmap='gray')
plt.axis('off')
plt.title('Decompressed Image')
plt.show()
```

```python
k_values = [10, 20, 50, 100]  # Example k values

plt.figure(figsize=(15, 10))

for i, k in enumerate(k_values):
    components = eigen_vectors[:, :k]
    compressed_data = np.dot(data_centered, components)
    decompressed_data = np.dot(compressed_data, components.T) + mean

    # Display the results
    plt.subplot(2, 2, i + 1)
    plt.imshow(decompressed_data, cmap='gray')
    plt.axis('off')
    plt.title(f'Decompressed Image (k = {k}, EVR = {np.sum(explained_variance_ratio[:k]):.2f})')

plt.show()
```

First, we sorted the eigenvectors in decreasing order according to their corresponding eigenvalues, therefore determining the main components. The main components bearing the information about the greatest image variance are the top eigen vectors.

Analyzing the cumulative explained variance plot helped us to ascertain the ideal number of main components for picture reconstruction.


We may see the information carried by varying component counts to rebuild the picture.

* Ten main components: The picture is created from a crude skeleton of the original image with only necessary information.

* 20 main components: Though the picture has blurriness, objects and features in it are clear.

* 50 main components: With most information kept from the original picture, the image is almost exactly like it.

* 100 primary vectors: The original picture and the rebuilt image are practically indistinguisable.




Principal Component Analysis may therefore help to reduce dimensionality in photos by maintaining most of the information while Enough 50 to 100 components in the above picture will allow one to recreate images without any obvious loss.
